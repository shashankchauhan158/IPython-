# Write a python program to display all the header tags from wikipedia.org

```python
# Python program to print all heading tags
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(url.text, 'html.parser')
story = soup.find_all(['h1', 'h2','h3'])
for i in story:
    print(i)
```

# Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
IMDB=requests.get("https://www.imdb.com/list/ls091520106/")
IMDB
```

```python
IMDB_Top100=BeautifulSoup(IMDB.content)
```

```python
IMDB_Name=[]
for b in IMDB_Top100.find_all("h3" , class_="lister-item-header"):
    IMDB_Name.append(b.text.replace('\n','').split("(")[0])
IMDB_Name
```

```python
IMDB_rating=[]
for a in IMDB_Top100.find_all("div" , class_="ipl-rating-star small"):
    IMDB_rating.append(a.text.replace('\n',''))
IMDB_rating
```

```python
IMDB_Year=[]
for c in IMDB_Top100.find_all("span" , class_="lister-item-year text-muted unbold"):
    IMDB_Year.append(c.text.replace("I",''))
IMDB_Year
```

```python
print(len(IMDB_Name),len(IMDB_rating),len(IMDB_Year))
```

```python
#making dataframe
import pandas as pd
df=pd.DataFrame({'Names':IMDB_Name,'Year of Release':IMDB_Year,'Ratings':IMDB_rating})
df
```

# Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm

```python
Presidents=requests.get("https://presidentofindia.nic.in/former-presidents.htm")
Presidents
```

```python
Presidents_list=BeautifulSoup(Presidents.content)
Presidents_list
```

```python
Presidents_Name=[]
for N in Presidents_list.find_all("div" , class_="presidentListing"):
    Presidents_Name.append(N.find("h3").text.split("(")[0])
Presidents_Name
```

```python
Presidents_TOS=[]
for T in Presidents_list.find_all("div",class_="presidentListing"):
    Presidents_TOS.append(T.find_all("p")[0].text.split(":")[1])
Presidents_TOS
```

```python
#making dataframe
import pandas as pd
President=pd.DataFrame({'Names':Presidents_Name,'Term of Service':Presidents_TOS})
President
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :

i) Restaurant name

ii) Cuisine

iii) Location

iv) Ratings

v) Image URL

```python
D=requests.get("https://www.dineout.co.in/delhi-restaurants/welcome-back")
D
```

```python
dineout=BeautifulSoup(D.content)
```

```python
#Restaurant name
R_name=[]
for i in dineout.find_all('div',class_="restnt-info cursor"):
    R_name.append(i.text.split("(")[0])
R_name
```

```python
price=[]
for i in dineout.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.split("(")[0])
price
```

```python
Location=[]
for i in dineout.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
Location
```

```python
Ratingss=[]
for i in dineout.find_all('div',class_="restnt-rating rating-4"):
    Ratingss.append(i.text)
Ratingss
```

```python
image_url=[]
for i in dineout.find_all('img',class_="no-img"):
    image_url.append(i['data-src'])
image_url
```

```python
import pandas as pd
Dineout=pd.DataFrame({"Restaurant Name":R_name, "Price":price , "Location":Location , "Ratings":Ratingss , "Images_Url":image_url})
Dineout
```
